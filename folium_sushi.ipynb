```python
스시 = pd.read_csv('list.csv', encoding='utf-8')
스시.tail()
```

```python
m = folium.Map(
    location = [37.5666, 126.978],
    tiles = 'openstreetmap',
    zoom_start = 11.5
)

tooltip = 'Click!!'

for i in range(스시.shape[0]):
    folium.Marker(
        [스시.iloc[i]['Latitude'], 스시.iloc[i]['Longitude']],
        popup = f'<div style="width:100px"><strong>{스시.iloc[i]["이름"]}</strong><br>\
        분류 : {스시.iloc[i]["분류"]}<br>\
        런치 : {스시.iloc[i]["런치"]}<br>\
        디너 : {스시.iloc[i]["디너"]}<br>\
        예약 : {스시.iloc[i]["예약"]}<br>\
        <a href="{스시.iloc[i]["링크"]}">상세페이지 이동</a></div>',
        tooltip = tooltip
    ).add_to(m)

m.save('index.html')
m
```
