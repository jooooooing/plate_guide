```python
스시 = pd.read_csv('list.csv', encoding='utf-8')
스시.head()
```

```python
m = folium.Map(
    location = [37.5666, 126.978],
    zoom_start = 12
)

엔트리급 = folium.FeatureGroup(name = '엔트리급').add_to(m)
미들급 = folium.FeatureGroup(name = '미들급').add_to(m)
하이엔드급 = folium.FeatureGroup(name = '하이엔드급').add_to(m)

for i in range(스시.shape[0]):
    if 스시.iloc[i]["분류"] == "엔트리급":
        folium.Marker(
            [스시.iloc[i]['Latitude'], 스시.iloc[i]['Longitude']],
            popup = f'<div style="width:120px"><strong>{스시.iloc[i]["이름"]}</strong><br>\
            분류 : {스시.iloc[i]["분류"]}<br>\
            런치 : {스시.iloc[i]["런치"]}<br>\
            디너 : {스시.iloc[i]["디너"]}<br>\
            예약 : {스시.iloc[i]["예약"]}<br>\
            <a href="{스시.iloc[i]["링크"]}">상세페이지 이동</a></div>',
        ).add_to(엔트리급)
    elif 스시.iloc[i]["분류"] == "미들급":
            folium.Marker(
              [스시.iloc[i]['Latitude'], 스시.iloc[i]['Longitude']],
              popup = f'<div style="width:120px"><strong>{스시.iloc[i]["이름"]}</strong><br>\
              분류 : {스시.iloc[i]["분류"]}<br>\
              런치 : {스시.iloc[i]["런치"]}<br>\
              디너 : {스시.iloc[i]["디너"]}<br>\
              예약 : {스시.iloc[i]["예약"]}<br>\
              <a href="{스시.iloc[i]["링크"]}">상세페이지 이동</a></div>',
        ).add_to(미들급)
    else:
        folium.Marker(
              [스시.iloc[i]['Latitude'], 스시.iloc[i]['Longitude']],
              popup = f'<div style="width:120px"><strong>{스시.iloc[i]["이름"]}</strong><br>\
              분류 : {스시.iloc[i]["분류"]}<br>\
              런치 : {스시.iloc[i]["런치"]}<br>\
              디너 : {스시.iloc[i]["디너"]}<br>\
              예약 : {스시.iloc[i]["예약"]}<br>\
              <a href="{스시.iloc[i]["링크"]}">상세페이지 이동</a></div>',
        ).add_to(하이엔드급)

folium.LayerControl(collapsed=False).add_to(m)

m.save('index.html')
m
```
